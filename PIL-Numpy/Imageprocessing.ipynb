import libraries

```python
from PIL import Image
from PIL import ImageOps
import numpy as np
import matplotlib.pyplot as plt #used to show image
```

```python
image = Image.open("NVG_green.png") #loading the images
```

```python
plt.imshow(image)
print(image)
```

Grey Scale Convertion

```python
Grey_image=ImageOps.grayscale(image)
Grey_image
```
